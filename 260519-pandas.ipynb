```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 150

shop = pd.DataFrame({
    'customer': np.random.choice(
        ['김민준','이서현','박지우','최하은','정도윤'], n),
    'category': np.random.choice(
        ['의류','전자기기','식품','생활용품','뷰티'], n),
    'quantity': np.random.randint(1, 6, n),
    'price': np.random.choice(
        [15000,29000,45000,89000,120000], n),
    'rating': np.random.choice(
        [1,2,3,4,5], n,
        p=[0.05,0.10,0.20,0.35,0.30]),
    'returned': np.random.choice(
        [True, False], n,
        p=[0.1, 0.9])
})

shop.loc[
    np.random.choice(shop.index, 8, replace=False),
    'rating'
] = float('nan')

shop.to_csv('shop_orders.csv', index=False)

df = pd.read_csv('shop_orders.csv')

df.head()
```

```python
df['total_sales'] = df['quantity'] * df['price']
df.groupby('category')['total_sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('category')['returned'].mean().sort_values(ascending=False)
```

```python
df['rating'] = df.groupby('category')['rating'].transform(lambda x: x.fillna(x.mean()))
df['rating'].isnull().sum()
```
